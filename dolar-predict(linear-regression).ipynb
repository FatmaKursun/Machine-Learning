```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
```

```python
veri=pd.read_csv('2016dolaralis.csv')
```

```python
veri.head()
```

```python
X= veri['Gun'].values
y= veri['Fiyat'].values
```

```python
X= X.reshape(251,1)
y=y.reshape(251,1)
```

```python
plt.scatter(X,y)
```

```python
plt.plot(X,y)
```

```python
tahminlineer= LinearRegression()
```

```python
tahminlineer.fit(X,y)
```

```python
tahminlineer.predict(X)
```

```python
plt.plot(X,tahminlineer.predict(X),c="red")
```

```python
tahminpolinom =PolynomialFeatures(degree=8)
```

```python
Xyeni=tahminpolinom.fit_transform(X)
```

```python
polinommodel=LinearRegression()
polinommodel.fit(Xyeni,y)
polinommodel.predict(Xyeni)
```

```python
plt.plot(X,y,polinommodel.predict(Xyeni))
```
